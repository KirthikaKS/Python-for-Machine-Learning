```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")  #this a inbuilt dataset in seaborn
```

```python
df.head()

#tip - dependent feature (because this depends on other features)
# others are independent feature

```

### Correlation with Heatmap

 A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types.
 It is very important in Feature Selection

```python
# helps to find the inter relationship between the feature (only when int or float , not possible for categorical type) - value ranges from -1 to +1

df.corr()
```

```python
sns.heatmap(df.corr())  #heatmap needs the matrix of correlation as input
```

## JoinPlot

A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot
### Univariate Analysis / Bivariate Analysis



```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')  #kind is the structure of the datapoints
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg') #draws the regression line - probability density function
```

## Pair plot

A pairs plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
#when you have more than two features, pairplot does the permutation and combination for everything

sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')  #hue is for categorizing the plots
```

## Dist plot

Dist plot helps us to check the distribution of the columns feature


```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)   #kde- kernel density estimation
```

## Categorical Plots

Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

- boxplot
- violinplot
- countplot
- bar plot

```python
## Count plot

sns.countplot('sex', data = df, )
```

```python
## Count plot

sns.countplot('day', data=df)
```

```python
## Bar plot

sns.countplot(y='sex', data = df)
```

```python
## Bar plot - needs both x and y values , where as countplot one axis is always count so giving one axis value is enough

sns.barplot(x='total_bill',y='sex', data = df)
```

```python
sns.barplot(x='sex',y='total_bill', data = df)
```

```python
sns.barplot(x='smoker',y='total_bill', data = df)
```

## Box plot

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot('sex','total_bill', data = df)
```

```python
sns.boxplot('day','total_bill',data = df,palette = 'rainbow')
```

```python
sns.boxplot(data = df, orient = 'v') #all features are taken
```

```python
# categorize my data based on some other categories

sns.boxplot('total_bill','day',hue = 'smoker', data = df)
```

## Violin Plot

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot('total_bill','day',data = df,palette = 'rainbow')
```
