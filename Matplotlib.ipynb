## MatplotLib

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

```python
import matplotlib.pyplot as plt

%matplotlib inline  # we dont have to write plt.show we can directlu run to get graph
```

```python
import numpy as np
```

```python
## Simple Examples

x= np.arange(0,10)
y=np.arange(50,60)

```

```python
##plotting using matplotlib 

##plt scatter

plt.scatter(x,y,c='b')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')

#this image can be save by right click or by the below fn

plt.savefig('2D graph.png')
```

```python
y=x*x
```

```python
## plt plot

plt.plot(x,y,'r*',linestyle = 'dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Graph')
```

```python
## Creating Subplots

plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'g--')
plt.subplot(2,2,4)
plt.plot(x,y,'bo')
```

```python
x = np.arange(1,11)
y = 3 * x+5
plt.title('Matplotlib Demo')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x,y,'--o',linewidth = 2,markersize = 8)
```

```python
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 4*np.pi,0.1)
y = np.sin(x)
plt.title("sine wave form")

# Plot the points using matplotlib 
plt.plot(x,y)
```

```python
#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 4*np.pi,0.1)
y_sine = np.sin(x)
y_cos = np.cos(x)
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2,1,1)

# Make the first plot 
plt.plot(x,y,'r--')
plt.title('Sine wave')

# Set the second subplot as active, and make the second plot. 
plt.subplot(2,1,2)
plt.plot(x,y,'g*')
plt.title('Cos wave')

# Show the figure. 
plt.show()
```

```python
## Bar plot

x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

## Histograms

## Box Plot using Matplotlib

```python
# helps us to percentiles

data = [np.random.normal(0,std,100) for std in range(1,4)]

# rectangular box plot
#default is vertical , if u say false for vertical it will be displayed as horizontal
#patch_artist fills the boxplot with color

plt.boxplot(data, vert = True, patch_artist = True)
```

## Pie Chart

```python
# Data to plot
labels = 'Python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.2,0,0,0)# explode 1st slice

# Plot
plt.pie(sizes,explode = explode, labels =labels,colors =colors,
autopct = '%1.1f%%', shadow = True)

plt.axis('equal')
plt.show()
```
